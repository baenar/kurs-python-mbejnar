```python
# Ocena podanej temperatury w skali 'Bardzo zimno' -> 'Bardzo ciepło'

temperatura = int(input("Podaj dzisiejszą temperaturę: "))

if (temperatura < -10):
    print('Bardzo zimno!')
elif (temperatura < 10):
    print('Zimno!')
elif (temperatura < 30): 
    print('Ciepło!')
else:
    print('Bardzo ciepło!')
```

```python
# Napisz funkcje sumującą elementy będące liczbami całkowitymi z danej listy

L_1 = [4.0, 'A', -2, 'text', 3.5, 1**7, (10,), [1, 2, 3], 0xA]

result = 0
for e in L_1:
    if type(e) == int:
        result += e
    elif type(e) == float and e % 1 == 0:
        result += e
print(int(result))
```
